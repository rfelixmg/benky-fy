```python
import pandas as pd
```

```python
verbs = pd.read_json("../datum/verbs.json")
verbs.head()
```

```python
df = pd.DataFrame(verbs[["conjugations"]]['conjugations'].to_list())
df.dictionary = df.dictionary.apply(lambda x: x['kanji'])
for key in ['polite', 'negative', 'polite_negative', 'past', 'polite_past', 'past_negative', 'te_form', 'potential', 'volitional']:
    try:
        df[key] = df[key].apply(lambda x: str(x['hiragana']))
    except:
        print("Error: ", key)
        pass
```
